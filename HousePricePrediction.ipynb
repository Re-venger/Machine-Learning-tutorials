# Implementation Using simple linear regression

##### Dataset used - https://www.kaggle.com/datasets/yasserh/housing-prices-dataset/data

1. Implemented simple linear regression using the area as a data feature to predict the price of the house.
2. low r2_score is because the data feature is not inherently dependent upon the area alone hence only 22% variance in prices can be explained.
3. in order to get the most acurate results we should use multiple linear regression.

#### This is implemented in the next Code - HousePricePridiction_MultiLR

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Housing.csv")
df = df.drop(columns=["bathrooms", "stories", "mainroad", "guestroom", "basement","hotwaterheating", "airconditioning", "parking", "prefarea", "bedrooms", "furnishingstatus"])
df.head()
```

```python
x = df.iloc[:, 1].values
y = df.iloc[:, 0:1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
x_test = x_test.reshape(-1, 1)
x_train = x_train.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
plt.scatter(x_train, y_train)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, lr.predict(x_test), color="red")
```

```python
from sklearn.metrics import r2_score
y_pred= lr.predict(x_test)
r2_score(y_test, y_pred)
```
