```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('Housing.csv')
```

```
# This is to reorder the cols to get prices as the y (O/P)
cols = list(df.columns)
cols = cols[1:] + [cols[0]]
df = df[cols]
df.head()

# checking if null vals exists or not
df.isnull().sum()
```

```
# Binary encoding the yes/no cols
binary_encode = {'yes':1, "no":0}
df[['mainroad', "guestroom", "basement", "hotwaterheating", "airconditioning", "prefarea"]] = df[['mainroad', "guestroom", "basement", "hotwaterheating", "airconditioning", "prefarea"]].replace(binary_encode)
df.head()
```

```
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

```

```
# encoding the non numberical data
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[
    ("tnf", OneHotEncoder(sparse_output=False, drop="first"), ['furnishingstatus'])
], remainder="passthrough")
x_train = ct.fit_transform(x_train)
x_test = ct.transform(x_test)
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```
y_pred = lr.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lr, x_train, y_train, cv=10)
scores.mean()
```
